```python
#calculation for net sales

def calculate_net_sales(sim, event_factors, class_share, product_share, gross_price, gtn):
    sim['Net Sales'] = (sim['Final Baseline Trend'] * (1 + event_factors) * class_share * product_share * gross_price * (1 - gtn))
    return sim
```

# Monte Carlo Simulation

This section implements a Monte Carlo simulation that randomly samples from the upside and downside parameters to generate multiple scenarios and calculate revenue distributions.

```python
def sample_from_distribution(dist_type, low=None, high=None, mean=None, std=None, left=None, mode=None, right=None):
    if dist_type == 'uniform':
        return np.random.uniform(low, high)
    elif dist_type == 'normal':
        return np.random.normal(mean, std)
    elif dist_type == 'triangular':
        return np.random.triangular(left, mode, right)
    elif dist_type == 'lognormal':
        return np.random.lognormal(mean, std)
    else:
        raise ValueError(f"Unsupported distribution type: {dist_type}")
    

# we need descrte values also
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Store sampled values
event_factor_list = []
class_share_list = []
product_share_list = []
gross_price_list = []
gtn_list = []
results = []

n_simulations = 100_000

for i in range(n_simulations):
    # Sample event factors
    event_factors = 0
    for j in range(1, 16):
        downside = globals()[f'DownsideEvent{j}']
        upside = globals()[f'UpsideEvent{j}']
        event_factors += sample_from_distribution(
            dist_type='uniform',
            low=downside, high=upside,
            mean=(downside + upside) / 2,
            std=(upside - downside) / 6,
            left=downside, mode=(downside + upside) / 2, right=upside
        )
    event_factor_list.append(event_factors)

    # Sample other parameters
    class_share = sample_from_distribution(
        dist_type='uniform',
        low=DownsideClassShare, high=UpsideClassShare,
        mean=(DownsideClassShare + UpsideClassShare) / 2,
        std=(UpsideClassShare - DownsideClassShare) / 6
    )
    class_share_list.append(class_share)

    product_share = sample_from_distribution(
        dist_type='normal',
        low=DownsideProductShare, high=UpsideProductShare,
        mean=(DownsideProductShare + UpsideProductShare) / 2,
        std=(UpsideProductShare - DownsideProductShare) / 6
    )
    product_share_list.append(product_share)

    gross_price = sample_from_distribution(
        dist_type='uniform',
        low=DownsideGrossPrice, high=UpsideGrossPrice,
        mean=(DownsideGrossPrice + UpsideGrossPrice) / 2,
        std=(UpsideGrossPrice - DownsideGrossPrice) / 6
    )
    gross_price_list.append(gross_price)

    gtn = sample_from_distribution(
        dist_type='uniform',
        low=DownsideGTN, high=UpsideGTN,
        mean=(DownsideGTN + UpsideGTN) / 2,
        std=(UpsideGTN - DownsideGTN) / 6
    )
    gtn_list.append(gtn)

    # Calculate Net Sales
    net_sales = calculate_net_sales(sim.copy(), event_factors, class_share, product_share, gross_price, gtn)['Net Sales'].iloc[0]
    results.append(net_sales)

# Convert results to numpy
results = np.array(results)

# Summary statistics
mean_sales = np.mean(results)
median_sales = np.median(results)
std_sales = np.std(results)
p5 = np.percentile(results, 2.5)
p95 = np.percentile(results, 97.5)

print(f"Simulations: {n_simulations}")
print(f"Mean Net Sales: {mean_sales:,.2f}")
print(f"Median Net Sales: {median_sales:,.2f}")
print(f"Std Dev: {std_sales:,.2f}")
print(f"2.5th Percentile: {p5:,.2f}")
print(f"97.5th Percentile: {p95:,.2f}")
```

```python
# Helper for plotting
def plot_distribution(data, name):
    plt.figure(figsize=(6, 4))
    sns.histplot(data, kde=True, bins=50)
    plt.title(f"{name} Distribution")
    plt.xlabel(name)
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Plot each

#plot individual events
plt.figure(figsize=(17, 15))
for i, (event, downside, upside) in enumerate(zip(range(1, 16), [globals()[f'DownsideEvent{j}'] for j in range(1, 16)], [globals()[f'UpsideEvent{j}'] for j in range(1, 16)])):
    plt.subplot(5, 3, i + 1)
    sns.histplot([sample_from_distribution('uniform', low=downside, high=upside) for _ in range(n_simulations)], kde=True, bins=50)
    plt.title(f'Event {event} Distribution')
    plt.xlabel('Event Factor')
    plt.ylabel('Frequency')
    plt.grid(True)

plt.tight_layout()
plt.suptitle('Distributions of Event1 to Event15 (Uniform)', fontsize=18, y=1.03)
plt.show()
plot_distribution(class_share_list, "Class Share")
plot_distribution(product_share_list, "Product Share")
plot_distribution(gross_price_list, "Gross Price")
plot_distribution(gtn_list, "GTN")
```
